```python
#2nd question : Write a Python Program for Pre order , Post order and In order Traversal.
#Creating the desired Tree
class Tree(object):
    def __init__(self,data):
        self.data=data
        self.left=None
        self.right=None
def preorder(root):
    if root:
        print(root.data)
        preorder(root.left)
        preorder(root.right)
def inorder(root):
    if root:
        inorder(root.left)
        print(root.data)
        inorder(root.right)
def postorder(root):
    if root:
        postorder(root.left)
        postorder(root.right)
        print(root.data)
#Creating the Tree(Ref Slide:160)
root=Tree('A')
root.left=Tree('B')
root.right=Tree('C')
root.left.left=Tree('D')
root.left.right=Tree('E')
root.left.right.left=Tree('M')
root.right.right=Tree('G')
root.right.left=Tree('F')
root.right.left.left=Tree('N')
print('Elements:')
print(root.data)
print(root.left.data)
print(root.right.data)
print(root.left.left.data)
print(root.left.right.data)
print(root.left.right.left.data)
print(root.right.right.data)
print(root.right.left.data)
print(root.right.left.left.data)
print('----------------------------')
print('Preorder:')
print(preorder(root))
print('----------------------------')
print('Inorder:')
print(inorder(root))
print('----------------------------')
print('Postorder:')
print(postorder(root))
#Output:
#Elements:
#A
#B
#C
#D
#E
#M
#G
#F
#N
#----------------------------
#Preorder:
#A
#B
#D
#E
#M
#C
#F
#N
#G
#None
#----------------------------
#Inorder:
#D
#B
#M
#E
#A
#N
#F
#C
#G
#None
#---------------------------
#Postorder:
#D
#M
#E
#B
#N
#F
#G
#C
#A
#None
```
